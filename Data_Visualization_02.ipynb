#### 1: List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#cause of deaths
cause = 'Chronic Diseases','Unintentional Injuries','Alzhemiers','Influenza and Pnuemonia','Sepsis','Others'
#percentile
percentile=[62,5,4,2,1,26]
```

#### 2: Draw a pie chart using the two data points

```python
#set the pie chart plot properties
```

```python
#set figure size
plt.figure(figsize=(10,10))
```

```python
#explode the largest pie in the data set
explode =(0.2,0,0,0,0,0.1)
```

```python
#set axis equal to draw pie as circle
plt.axis('equal')
```

```python
#set title of the pie chart
plt.title('Ohio State 2012 : Leading causes of deaths')
```

```python
#show the plot (pie chart)
plt.show()
```

```python
#set pie chart properties
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%',startangle=60)
```
